```python
#  加载数据处理包
import numpy as np   # pandas和numpy常常结合在一起使用，导入numpy库
import pandas as pd  # 导入pandas库
import os.path

print(pd.__version__) # 打印pandas版本信息
```

```python
resourse_data_folder = '/home/maocy/datasets/zip'
resourse_data_filename = 'DBpedia_duplicate.csv'
resourse_data = pd.read_csv(
    os.path.join(resourse_data_folder, resourse_data_filename),
    sep = '\t',
    names = ['subject', 'predicate', 'object']
)
print(f'resourse_data.shape {resourse_data.shape}')
```

```python
unique_predicate = np.unique(resourse_data['predicate'].values)
print(f'Total unique predicate in resourse_data: {len(unique_predicate)}')
```

```python
filter_in_predicates = ['http://www.georss.org/georss/point']

merged_df_some = resourse_data[resourse_data['predicate'].isin(filter_in_predicates)]
print(f'merged_df_some.shape: {len(merged_df_some)}')
```

```python
random_sample = merged_df_some.sample(n=1000, random_state=66)
random_sample
```

```python
random_sample[['lat', 'lon']] = random_sample['object'].str.split(expand=True)
random_sample
```

```python
random_sample['lat'] = random_sample['lat'].astype('float')
random_sample['lon'] = random_sample['lon'].astype('float')
random_sample
```

```python
import pygeohash as gh
def get_geohash(row):
    '''
    Calculate geohash
    '''
    return gh.encode(row['lat'], row['lon'], precision = 6)
```

```python
from tqdm import tqdm
tqdm.pandas()
random_sample['geohash'] = random_sample.progress_apply(get_geohash, axis = 1)
random_sample
```

```python
random_sample['1_precision'] = random_sample['geohash'].apply(lambda x: x[0])
random_sample['2_precision'] = random_sample['geohash'].apply(lambda x: x[:2])
random_sample['3_precision'] = random_sample['geohash'].apply(lambda x: x[:3])
random_sample['4_precision'] = random_sample['geohash'].apply(lambda x: x[:4])
random_sample['5_precision'] = random_sample['geohash'].apply(lambda x: x[:5])
random_sample['6_precision'] = random_sample['geohash'].apply(lambda x: x)
random_sample
```

```python
random_sample.to_csv("Georss random_sample.csv", sep = '\t', index = False)
```
